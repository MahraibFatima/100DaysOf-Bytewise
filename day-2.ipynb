## 4. Prime Number Checker

```python
def isPrime(num):
    if num <= 1:
        return False
    elif num == 2:
        return True
    else:
        for i in range(2, int(num**0.5) + 1):
            if (num % i) == 0:
                return False
        return True

num = int(input("Enter a number: "))
if(isPrime(num)):
    print(num, "is a prime number!")
else: 
    print(num, "is not a prime number!")
```

## 5. Guess the Number Game

- Write a program where the computer generates a random number, and the user has to guess it.


```python
import random
secret = random.randint(1, 100)

while(True):
    guess = int(input("Enter your guess in range(1,100): "))
    
    if guess < secret:
        print("Too low! Try a higher number.")
    elif guess > secret:
        print("Too high! Try a lower number.")
    else:
        print(f"Congratulations! You guessed the number {secret}!")
        break
print("Thanks for playing!")
```

## 7. Palindrome Sentences

```python
import string

def is_palindrome(sentence):
    normalized_sentence = ''.join(char.lower() for char in sentence if char.isalnum())
    return normalized_sentence == normalized_sentence[::-1]

sentence = input("Enter a sentence")
if(is_palindrome(sentence)):
    print("It's a plaindrome!")
else:
    print("It's not a plaindrome!")
```

## 8. Anagram Checker

```python
def isAragrams(s1, s2):
    if len(s1) != len(s2):
        return False
    
    count1 = {}
    count2 = {}
    
    for char in s1:
        count1[char] = count1.get(char, 0) + 1
    
    for char in s2:
        count2[char] = count2.get(char, 0) + 1
        
    return count1 == count2

s1 = input("Enter first word: ")
s2 = input("Enter second word: ")

s1 = s1.replace(" ", "").lower()
s2 = s2.replace(" ", "").lower()

if(isAragrams(s1, s2)):
    print("Words are Aragrams!")
else: 
    print("Words are not Aragrams!")
```

## 9. Reverse Words in a Sentence

```python
def reverse_words(sentence):
    words = sentence.split()
    reversed_words = words[::-1]
    reversed_sentence = ' '.join(reversed_words)
    
    return reversed_sentence

sentence = input("Enter a sentence: ")
print("Reversed sentence:", reverse_words(sentence))
```

## 10. Temperature Converter

```python
def celsius_to_fahrenheit(temp):
    return (temp * 9/5) + 32

def celsius_to_kelvin(temp):
    return temp + 273.15

temp = float(input("Enter temperatures: ") )
print("Celsius to Fahrenheit:",celsius_to_fahrenheit(temp))
print("Celsius to Kelvin: ", celsius_to_kelvin(temp))
```
