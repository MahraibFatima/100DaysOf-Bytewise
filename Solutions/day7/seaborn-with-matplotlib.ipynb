```python
import numpy as np
import seaborn as sns
import pandas as pd
import warnings
import matplotlib.pyplot as plt
```

## Line Plot

```python
values = [2, 5, 1, 8, 3, 7]
timestamps = range(len(values)) 

plt.plot(timestamps, values)
plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Trend of Values Over Time")

# Display the plot
plt.show()
```

## Bar Chart 

```python
data = pd.Series(["Cat", "Dog", "Bird", "Cat", "Rabbit", "Dog"])
category_counts = data.value_counts()

category_counts.plot(kind='bar')
plt.xlabel("Category")
plt.ylabel("Frequency")
plt.title("Frequency of Categories")
plt.show()
```

## Scatter Plot

```python
x_data = [1, 3, 5, 7, 2, 4]
y_data = [4, 6, 5, 8, 3, 7]

plt.scatter(x_data, y_data)
plt.xlabel("Variable 1")
plt.ylabel("Variable 2")
plt.title("Relationship Between Variables")
#optional
plt.scatter(x_data, y_data, marker='o', color='blue')

# Display the plot
plt.show()
```

## Pairplot

```python
#load toy dataset
tips = sns.load_dataset('tips')

warnings.simplefilter(action='ignore')
sns.pairplot(tips)
plt.show()
```

## Box Plot

```python
tips.head()
```

```python
sns.boxplot(x=tips["tip"])

plt.title("Distribution of Numerical Values by Category")
plt.xlabel("Category")
plt.ylabel("Numerical Value")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis

plt.show()
```

## Heatmap

```python
numerical_columns = ['total_bill', 'tip']
tips_filtered = tips[numerical_columns]
correlation_matrix = tips_filtered.corr()

# Create a heatmap
sns.heatmap(correlation_matrix, annot=True)

plt.title("Correlation Matrix of Tips Dataset")
plt.show()
```

## Subplot Grid

```python
rows = 2
cols = 2

fig, axes = plt.subplots(rows, cols, figsize=(10, 6))

ax1 = axes[0, 0]  # Top-left subplot
sns.scatterplot(x='total_bill', y='tip', data=tips, ax=ax1)
ax1.set_title('Tip vs Total Bill')

ax2 = axes[0, 1]  # Top-right subplot
sns.boxplot(x='day', y='total_bill', data=tips, showmeans=True, ax=ax2)
ax2.set_title('Total Bill Distribution by Day')

ax3 = axes[1, 0]  # Bottom-left subplot
sns.distplot(tips['tip'], kde=True, ax=ax3)
ax3.set_title('Tip Distribution')

ax4 = axes[1, 1]  # Bottom-right subplot
sns.countplot(x='sex', data=tips, ax=ax4)
ax4.set_title('Count of Customers by Sex')

plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1, wspace=0.4, hspace=0.4)

plt.show()
```

## Customize the appearance of a Seaborn plot

```python
custom_palette = ["skyblue", "green"]
sns.scatterplot(
    x="total_bill",
    y="tip",
    data=tips,
    palette=custom_palette,  # Use custom color palette
    hue="sex",  
    style="day",  
    legend="full"
)

plt.title("Tip Amount vs Total Bill (Colored by Sex)")
plt.xlabel("Total Bill Amount (USD)")
plt.ylabel("Tip Amount (USD)")
plt.xticks(rotation=45)  
```

## Violin Plot

```python
sns.violinplot(
    x = "day",  
    y = "tip",  
    showmeans=True,  
    data=tips
)

plt.title("Distribution of Tips by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Tip Amount (USD)")
plt.xticks(rotation=45)  

plt.show()
```

##  KDE plot on Histogram

```python
tips = tips["tip"]

fig, ax = plt.subplots(figsize=(8, 6))
sns.histplot(tips, kde=True, ax=ax, stat="density") 
plt.title("Distribution of Tip Amount (USD)")
plt.xlabel("Tip Amount")
plt.ylabel("Density")

plt.show()
```
